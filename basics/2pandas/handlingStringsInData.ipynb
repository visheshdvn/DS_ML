## Handling Strings in data

```python
import pandas as pd, numpy as np

iris = pd.read_csv("../../datasets/iris.csv")
print(type(iris))
df = iris.copy()
df.columns = ['sl', 'sw', 'pl', 'pw', 'flower_type']
```

```python
df["diff_pl_pw"] = df["pl"] - df["pw"]
df.head()
```

```python
df["Gender"] = "Female"
df.head()
```

```python
df.iloc[0:10, 6] = "Male"
df
```

```python
def f(s):
    if s == "Male":
        return 0
    return 1

df['sex'] = df.Gender.apply(f)
df
```

```python
del df["Gender"]
df.head()
```
